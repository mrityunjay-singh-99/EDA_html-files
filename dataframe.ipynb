```python
#create a empty dataframe and add value use list comprehension

import pandas as pd 
df2=pd.DataFrame()
l1=[i for i in range(10)]
l2=[i*2 for i in range(10)]
l3=[i*3 for i in range(10)]
df2["number"]=l1
df2["square"]=l2
df2["cube"]=l3

df2
```

```python
#update columns in existing table
l4=[i**4 for i in range(10)]
df2["cube"]=l4
df2
```

```python
#drop coulms in coulmns and row in exisiting table
col_op=df2.drop("cube",axis=1)   #for column axis =1
row_op=df2.drop(9,axis=0)        #for column axis=0
col_op
```

```python
row_op
```

```python
# how to remove two and more columns and row

col_op=df2.drop(["cube","square"],axis=1)   #for column axis =1
row_op=df2.drop([9,5],axis=0)        #for column axis=0
col_op
```
