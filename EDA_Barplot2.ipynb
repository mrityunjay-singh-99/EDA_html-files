```python
bank_df=pd.read_csv("bank.csv",sep=";")   #read datasets
bank_df     
```

```python
bank_df.dtypes  #check datatypes
```

```python
 #make different list for categorical and numerical

cat_col=[keys for keys,value in dict(bank_df.dtypes).items() if value=="object"]    
num_col=[keys for keys,value in dict(bank_df.dtypes).items() if value!="object"]
num_col
```

```python
cat_col
```

```python
len(cat_col)
```

```python
bank_df["job"].value_counts().keys()    #
```

```python
#plot bank categorical columns using subplot

plt.figure(figsize=(22,15))
for i in range(len(cat_col)):
    plt.subplot(4,4,i+1)
    sns.countplot(data=bank_df,
                 x=cat_col[i],
                 order=bank_df[cat_col[i]].value_counts().keys())
    plt.xlabel(f"{cat_col[i]}")
    plt.ylabel("class frequency")
    
```

```python
# plot bar chart

dnf=bank_df["job"].value_counts()
ax=dnf.plot(kind="bar")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# plot bar charts

for i in cat_col:
    dnf=bank_df[i].value_counts()
    ax=dnf.plot(kind="bar")
    ax.bar_label(ax.containers[0])
    plt.show()
```

```python
bank_df["job"].value_counts()
```

```python
#plot pie charts
dnf=bank_df["job"].value_counts()
keys=dnf.keys()
value=dnf.values
plt.pie(x=value,
        labels=keys,
       autopct="%0.2f%%",
       radius=2,
       explode=[1,0.1,1,0.1,1,0.1,1,0.1,1,0.1,0.1,1])
plt.show()
```
