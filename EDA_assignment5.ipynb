```python
l=[1,2,3,4,5,13]

for i in num_colm:
    data=bank_df[i]
    n=len(data)
    
    
    for k in l:
        if n%2**k==n:
            plt.hist(data,bins=k)
            plt.xlabel(f"{i}")
            plt.ylabel("class frequency")
        plt.show()
```

```python
for i in num_colm:
    data=bank_df[i]
    n=len(data)
        
        
    k=0
    while True:
        k+=1
        if n%2**k==n:
            plt.hist(data,bins=k)
            plt.xlabel(f"{i}")
            plt.ylabel("class frequency")
            break
        plt.show()
        
```
